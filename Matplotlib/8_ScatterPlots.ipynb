```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Gerando um dataset ficticio sobre imigrantes para o Canadá

df = pd.DataFrame({'Years':[2009,2010,2011,2012,2013],'Total':[100,500,1000,1500,2000]})
df.head
```

<h2>Scatter Plots / Gráfico de Dispersão</h2>
<p>
Os gráficos de dispersão ou Scatter plot são representações gráficas do relacionamento entre duas variáveis numéricas. O Scatter plot utiliza pontos para representar essa relação, cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical.
</p><br>


```python
df.plot(kind='scatter', x='Years',y='Total') # X = Coluna que ficará no eixo X, e Y coluna que ficará no eixo Y
plt.title('Total Imigrantes')
plt.xlabel('Anos')
plt.ylabel('Número Imigrantes')
plt.show()
```

```python
ax = df.plot(kind='scatter', x='Years',y='Total')
ax.set_title('Total Imigrantes')
ax.set_xlabel('Anos')
ax.set_ylabel('Número imigrantes')
plt.show()
```

Note que os gráficos tem uma tendência ascendente podemos analisar essa tendencia traçando uma linha<br>
Vamos traçar uma linha pra tentar prever o número de imigrantes em 2015

```python
# Para isso precisamos obter a equação de melhor ajuste com o numpy 'polyfit()'
x = df['Years']
y = df['Total'] 
fit = np.polyfit(x,y,deg=1)
fit
```

A saida é um array com os coeficiêntes polinomiais traçando uma regressão linear(y=a*x+b), sendo o indice 0 a inclinação
e o 1 a interceptação

```python
df.plot(kind='scatter', x='Years',y='Total') # X = Coluna que ficará no eixo X, e Y coluna que ficará no eixo Y
plt.title('Total Imigrantes')
plt.xlabel('Anos')
plt.ylabel('Número Imigrantes')


plt.plot(x, fit[0] * x + fit[1], color='red') # Lembre-se que x é os anos e fit é o array da regressão linear
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(0, 150000))

plt.show()
```
