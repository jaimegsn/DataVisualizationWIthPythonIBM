<h2>Gráfico de Linha</h2>
<p>
Gráficos de linhas mostram tendências ou alterações ao longo do tempo exibindo uma série de pontos de dados conectados por segmentos de linha reta. Você pode exibir um único campo de dados como várias linhas com base em diferentes categorias por meio de Cor por campo.
</p><br>

<h3>Quando Usar ?</h3>
<p>
O melhor caso de uso para um gráfico de linhas é quando você tem um conjunto de dados contínuo e seu objetivo é visualizar como os dados se comportam durante um período de tempo.
</p>

```python
years = list(map(str,range(2009,2014))) # lista de 2009 á 2014 em formato de str armazenado na variavel years
processData.loc['Brazil',years].plot(kind='line') # especificando colunas de acordo com a lista
plt.title('Imigrantes do Brasil')
plt.ylabel('Numeros de Imigrantes')
plt.xlabel('Anos')
plt.show()
```
