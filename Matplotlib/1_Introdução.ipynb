```python
import matplotlib.pyplot as plt
import numpy as np
```

Fazendo um simples gráfico onde irá gera circulo nas cornadas 5(y) e 5(x)

```python
plt.plot(5,5,'o')
plt.ylabel('Y') # Definindo nome do eixo Y
plt.xlabel('X') # Definindo nome do eixo X
plt.title('Definindo titulo do gráfico')
plt.show()
```

O pandas possui integração com o matplotlib direto de um DataFrame ex:

```python
import pandas as pd
```

```python
imigrantes = pd.DataFrame({'Year':['1980','1981','1982'],'India':[8000,2999,1210],'China':[1736,3723,13789]})
imigrantes = imigrantes.set_index('Year') # Pode se usar um pd.Series também e a propriedade append para manter 2 indices o 2,3,4 e o ano
imigrantes.head()
```

```python
imigrantes.plot(kind='line')  # Chamamos plot e dentro do parâmetro kind especificamos o tipo do gráfico
```

```python
imigrantes.plot(kind='bar')  # Grafico de Barras
```
